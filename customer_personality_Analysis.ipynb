**5. Handling the Missing values**

There are two methods for filling missing values in the columns


*   By fill missing values for integars

    df['column1'].fillna(df['column1'].mean(), inplace=True)
    
    
*   For strings we will utilize the mode() method to fill missing values

    df['column1'].fillna(df['column1'].mode(), inplace=True)




**Method for Integars**

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(), inplace=True)
```

**Method for Strings**

```python
df['Outlet_Size'].fillna(df['Outlet_Size'].mode().iloc[0], inplace=True)
```


**Central tendency** It provides a summary

measure that gives an indication of the center of a distribution.

The three stimated parameters of central tendency are:

**Mean:** The mean, or average, is the sum of all values in a dataset divided by the total number of values.

**Median:** The median is the average of the two middle values to its left or right, when data is in arranged form.

**Mode:** The mode is the value that appears most frequently in a dataset.
